```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
from pandas import ExcelWriter
from pandas import ExcelFile

myRetailData = pd.read_excel('Online Retail.xlsx')
```

```python
myRetailData.head()
```

```python
myRetailData['Description'] = myRetailData['Description'].str.strip()
myRetailData.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
myRetailData['InvoiceNo'] = myRetailData['InvoiceNo'].astype('str')
myRetailData = myRetailData[~myRetailData['InvoiceNo'].str.contains('C')]
myRetailData.head()
```

```python
myRetailData['Country'].value_counts()
```

```python
myRetailData.shape
```

```python
print(myRetailData.columns)
```

```python
SpainBasket = (myRetailData[myRetailData['Country'] == "Spain"].groupby(['InvoiceNo', 'Description'])['Quantity']
               .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
SpainBasket.head()
```

```python
def function(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
SpainBasket_sets = SpainBasket.applymap(function)
SpainBasket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
SpainBasket_sets.head()
```

```python
Popular_itemsets = apriori(SpainBasket_sets.astype('bool'), min_support=0.06, use_colnames=True)
```

```python
Spain_rules = association_rules(Popular_itemsets, metric='lift', min_threshold=1)
```

```python
Spain_rules.head(100)
```

```python
sorted_spain_rules_support = Spain_rules.sort_values(by='support', ascending=False)
sorted_spain_rules_support.head()
```

```python
sorted_spain_rules_confidence = Spain_rules.sort_values(by='confidence', ascending=False)
sorted_spain_rules_confidence.head()
```
